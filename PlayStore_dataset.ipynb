# SubTask2: Removing Irrevalant Values

To get a list of unique values Use .unique() function. If thos function does not comply with te data type of the column or does not
justify the characteristic of the attribute, then remove all the rows having that unique value.

```python
# Checking the unique values in the Category column
apps_df['Category'].unique()
```

```python
# Droping the row with 1.9
apps_df= apps_df[apps_df['Category']!='1.9']
apps_df
```

```python
# Checking the unique values in the Category column
apps_df['Category'].unique()
```

```python
apps_df['Rating'].unique()
```

```python
# Replacing null value to zero
apps_df= apps_df.replace(np.nan, 0)
```

```python

apps_df['Rating'].value_counts()
```

```python
#apps_df.drop("Life Made WI-Fi Touchscreen Photo Frame", inplace=True)
```

```python
apps_df['Type'].value_counts()
```

```python
apps_df= apps_df[apps_df['Type']!=0]
```

```python
apps_df['Type'].value_counts()
```

```python
#Droping the price value less than 0 and 0
apps_df= apps_df[apps_df['Price']>=0]
```

```python
apps_df['Content Rating'].value_counts()
```

```python
# dropping the value Unrated and Everyone 10+
apps_df= apps_df[apps_df['Content Rating']!='Unrated']
apps_df= apps_df[apps_df['Content Rating']!='Everyone 10+']
```

```python
apps_df['Content Rating'].value_counts()
```

```python
apps_df['Genres'].unique()
```

```python
# Education and Educational values are conflicting in this column
#So merged this two values
apps_df['Genres']= apps_df['Genres'].str.replace ('Educational', 'Education')

#Confirming the merged value
apps_df['Genres'].unique()
```

```python
apps_df['Last Updated'].unique()
```

```python
apps_df['Current Ver'].unique()
```

```python
# Fill the nan value with blanks in order to ignore value error
app_ver=apps_df['Current Ver'].replace(np.nan, '')

#Filter for rows that contains only numeric value
apps_df=apps_df[~app_ver.str.contains(r'\s', r'^a-zA-Z',na=False)]


#Confirm Changes
list(apps_df['Current Ver'].unique())
```

```python
apps_df['Android Ver'].unique()
```

```python
#Dropping rows that contain nan and a specific value
apps_df= apps_df[(apps_df['Android Ver'].notnull()) & (apps_df['Android Ver'] != 'Varies with device')]

#Confirm changes
apps_df['Android Ver'].unique()
```

```python
# Removing duplicates for review dataset

reviews_df['Translated_Review'].unique()
```

```python
#Dropping null value
reviews_df=reviews_df[reviews_df['Translated_Review'].notnull()]
reviews_df['Translated_Review'].unique()
```

```python
reviews_df['Sentiment'].unique()
```

```python
# Dropping null values
reviews_df=reviews_df[reviews_df['Sentiment'].notnull()]
reviews_df['Sentiment'].unique()
```

```python
reviews_df['Sentiment_Polarity'].unique()
```

```python
# Dropping null values
reviews_df=reviews_df[reviews_df['Sentiment_Polarity'].notnull()]
reviews_df['Sentiment_Polarity'].unique()
```

```python
reviews_df['Sentiment_Subjectivity'].unique()
```

```python
#Droping null values
reviews_df=reviews_df[reviews_df['Sentiment_Subjectivity'].notnull()]
reviews_df['Sentiment_Subjectivity'].unique()
```
